```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
titanic = sns.load_dataset('titanic')

```

```python
titanic.head()
```

## 수치형 데이터 시각화

- **수치형 데이터 시각화의 종류 4가지**

    - histplot()

    - kdeplot()

    - displot()
    
    - rugplot()

```python
fig, ax = plt.subplots(figsize = (10, 4))
sns.histplot(data=titanic, x = 'age', bins = 10, hue = "alive", multiple="stack")
ax.set_title("Age Histogram")
ax.set_xlabel("Age")
ax.set_ylabel("Freq")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 4))
sns.kdeplot(data=titanic, x = 'age', hue="alive", multiple="stack")
ax.set_title("Age KDE")
ax.set_xlabel("Age")
ax.set_ylabel("Freq")
plt.show()
```

```python
sns.displot(data=titanic, x='age')
# sns.displot(data=titanic, x='age', kind='kde')
# sns.displot(data=titanic, x='age', kde = True)
```

```python
fig, ax = plt.subplots(figsize = (10, 4))
sns.kdeplot(data=titanic, x = 'age')
sns.rugplot(data=titanic, x = 'age') # 값의 밀집여부를 보여줌.
ax.set_title("Age KDE")
ax.set_xlabel("Age")
ax.set_ylabel("Freq")
plt.show()
```

## 범주형 데이터 시각화

- **범주형 데이처 시각화 6가지**

    - barplot()
    
    - pointplot()

    - boxplot(), violinplot()

    - countplot()

    - pieplot()

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.barplot(x = 'class', y='fare', data = titanic)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.pointplot(x = 'class', y='fare', data = titanic)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.boxplot(x = 'class', y='fare', data = titanic)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.violinplot(x = 'class', y='fare', hue='sex', split = True, data = titanic)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.countplot(y = 'class', data = titanic)
plt.show()
```

```python
plt.pie(x=[10,60,30], labels = ['A','B','C'], autopct="%.1f%%")
```

## 데이터 관계 시각화

```python
flights = sns.load_dataset('flights')

flights.head()
```

```python
flights_pivot = flights.pivot(index='month',
                              columns='year',
                              values='passengers')

flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
```
